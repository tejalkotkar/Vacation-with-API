```python
import json
import requests
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import date

# Imporing API key
import sys
sys.path.append('..')
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy
```

## Generate Cities List

```python
lat_langs = []
cities = []

# Create a set of random lat and lang combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1600)
langs = np.random.uniform(low=-180.000, high=180.000, size=1600)

# Zip lat & lang 
lat_langs = zip(lats, langs)

for lat_lang in lat_langs:
    city = citipy.nearest_city(lat_lang[0],lat_lang[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create a dataframe to store values
city_weather = pd.DataFrame({"City" : cities, "Lat":"", "Lng" : "", "Max Temp" : "", "Humidity" : "",
                             "Cloudiness" : "", "Wind Speed" : "", "Country" : "",  "Date" : ""})
city_weather.head()
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?" + "&appid=" + weather_api_key + "&units=imperial"
recordsset = 1
recordnum = 1

print("Beginning Data Retrieval\n-----------------------------")
for index, row in city_weather.iterrows():
    city = row['City']
    query_url = url + "&q=" + city
    response = requests.get(query_url).json()
    try:
        print(f"Processing Record {recordnum} of Set {recordsset} | {response['name']}")
        city_weather.loc[index,'Lat'] = response['coord']['lat']
        city_weather.loc[index,'Lng'] = response['coord']['lon']
        city_weather.loc[index,'Max Temp'] = response['main']['temp_max']
        city_weather.loc[index,'Humidity'] = response['main']['humidity']
        city_weather.loc[index,'Cloudiness'] = response['clouds']['all']
        city_weather.loc[index,'Wind Speed'] = response['wind']['speed']
        city_weather.loc[index,'Country'] = response['sys']['country']
        city_weather.loc[index,'Date'] = response['dt']
    except(KeyError, IndexError):
        print("City not found. Skipping...")
    if recordnum <= 50:
        recordnum+=1
    else:
        recordsset+=1
        recordnum = 1
print("-----------------------------\nData Retrieval Complete\n-----------------------------")    


#response
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Replace blank values with NaN
city_weather = city_weather.replace("", np.nan, regex=True)
city_weather.count()
```

```python
# Drop blank records
city_weather = city_weather.dropna(how='any')

city_weather.count()
```

```python
# Save a clean data to csv
city_weather.to_csv("city_weather_data.csv", encoding="utf-8", index=False)
```

## Inspect the data and remove the cities where the humidity > 100%.
----
Skip this step if there are no cities that have humidity > 100%. 

```python
city_weather.describe()
```

```python
#  Get the indices of cities that have humidity over 100%.
humidity_list = city_weather.index[city_weather["Humidity"] > 100].tolist()
len(humidity_list)
```

```python
# Make a new DataFrame equal to the city data to drop all humidity outliers by index.
# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".
if len(humidity_list) > 0:
    print("In If")
    clean_city_data = city_weather.drop(city_weather.index[humidity_list], inplace=False)
else:
    print("In Else")
    clean_city_data = city_weather
```

```python
clean_city_data
```

## Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

## Latitude vs. Temperature Plot

```python
today = date.today()

plt.scatter(clean_city_data["Lat"], clean_city_data["Max Temp"], edgecolors="black")
plt.title("City Latitude vs. Max Temperature ({})".format(today.strftime("%m/%d/%y")))
plt.xlabel("Lattitude")
plt.ylabel("Max Temperature (F)")
plt.ylim(clean_city_data["Max Temp"].min()-10, clean_city_data["Max Temp"].max()+10)
plt.grid(color="grey")
plt.savefig("Plot_Images/1_Lat_Temp_Scatter.png")
plt.show()
```

## Latitude vs. Humidity Plot

```python
plt.scatter(clean_city_data["Lat"], clean_city_data["Humidity"], edgecolors="black")
plt.title("City Latitude vs. Humidity ({})".format(today.strftime("%m/%d/%y")))
plt.xlabel("Lattitude")
plt.ylabel("Humidity(%)")
plt.ylim(clean_city_data["Humidity"].min()-10, clean_city_data["Humidity"].max()+10)
plt.grid(color="grey")
plt.savefig("Plot_Images/1_Lat_Humidity_Scatter.png")
plt.show()
```
